###
## split
###

```python
import seaborn as sns
import matplotlib.pyplot as plt

a = sns.load_dataset('tips')
sns.catplot( x = 'total_bill', y = 'day', data = a, kind = 'violin', split = True, hue = 'sex') 
```

###
## violin + swarm plot together
###

```python

import seaborn as sns
import matplotlib.pyplot as plt

a = sns.load_dataset('tips')
g = sns.catplot(x='day', y='total_bill', data=a, kind='violin', split=True, hue='sex')
sns.swarmplot(x='day', y='total_bill', data=a, color='k', ax=g.ax)
```

###

```python

import seaborn as sns
import matplotlib.pyplot as plt

a = sns.load_dataset('tips')
g = sns.catplot(x='day', y='total_bill', data=a, kind='violin', split=True, hue='sex', color = 'c')
sns.swarmplot(x='day', y='total_bill', data=a, color='k', ax=g.ax)
```
