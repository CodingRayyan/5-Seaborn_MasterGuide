###
## Setting Colours for hue
###


```python

import seaborn as sns
import pandas as pd

tips = sns.load_dataset('tips')
tips

colours = { 'Lunch' : 'Blue' , 'Dinner' : 'Green'}

sns.relplot( x = 'total_bill', y = 'tip', data = tips, hue = 'time', palette = colours)
```

###
## Setting Style of Our Own Choice
###

```python

import seaborn as sns
import pandas as pd

a = sns.load_dataset('tips')

colours = ['blue','orange','green','red']
styles = {'Lunch' : 's' , 'Dinner' : 'o'}
sns.relplot( x = 'total_bill', y = 'tip', data = a, hue = 'day', palette = colours, style = 'time', markers = styles, size = 'sex', sizes = (15,100))
```

###
## Using col
col_wrap means no. of coulmns

###

```python

import seaborn as sns
import pandas as pd

a = sns.load_dataset('tips')

colours = ['blue','orange','green','red']
styles = {'Lunch' : 's' , 'Dinner' : 'o'}
sns.relplot( x = 'total_bill', y = 'tip', data = a, hue = 'day', palette = colours, style = 'time', markers = styles, size = 'sex', sizes = (15,100), col = 'smoker', col_wrap = 2)
```

```python

import seaborn as sns
import pandas as pd

a = sns.load_dataset('tips')

colours = ['blue','orange','green','red']
styles = {'Lunch' : 's' , 'Dinner' : 'o'}
sns.relplot( x = 'total_bill', y = 'tip', data = a, hue = 'smoker', palette = colours, style = 'time', markers = styles, size = 'sex', sizes = (15,100), col = 'day', col_wrap = 2)
```
