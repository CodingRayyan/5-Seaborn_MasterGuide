###
## jitter
###

```python

import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt

a = pd.read_csv('groupby(csvfile8).csv')
print(a)
sns.catplot( x = 'College', y = 'Height(foot.inch)', data = a, jitter = False)
```

###
## swarm plot
###

```python

import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt

a = pd.read_csv('groupby(csvfile8).csv')
print(a)
sns.catplot( x = 'College', y = 'Height(foot.inch)', data = a, kind = 'swarm')
```

###
## swarm plot (it doesn't support style & size)

```python

import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt

a = pd.read_csv('groupby(csvfile8).csv')
print(a)
sns.catplot( x = 'College', y = 'Height(foot.inch)', data = a, kind = 'swarm', hue = 'Rank')
```

###
## swarm plot (it doesn't support style & size)
###

```python

import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt

a = sns.load_dataset('tips')
print(a)
sns.catplot( x = 'smoker', y = 'tip', data = a, kind = 'swarm', hue = 'time')
```

###
## Passing order
###

```python

import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt

a = sns.load_dataset('tips')
print(a)
sns.catplot( x = 'smoker', y = 'tip', data = a, kind = 'swarm', hue = 'time', order = ['No','Yes'])
```
