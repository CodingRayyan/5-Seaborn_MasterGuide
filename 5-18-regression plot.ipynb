```python

import seaborn as sns

a = sns.load_dataset('tips')
a.head()
sns.regplot( x = 'total_bill', y = 'tip', data = a)
# the line and ci shows the probability that how the tips will increase with respect to totalbill
```

```python
# regplot

import seaborn as sns
import pandas as pd

a = pd.read_csv('groupby(csvfile8).csv')
a,sns.regplot( x = 'Rank', y = 'Height(foot.inch)', data = a)
# the line and ci shows the probability that how the height will increase or decrease with respect to rank
```

###
## lmplot 
it support hue, col, row

###

```python

import seaborn as sns
import pandas as pd

a = pd.read_csv('groupby(csvfile8).csv')
a,sns.lmplot( x = 'Rank', y = 'Height(foot.inch)', data = a, hue = 'Department')
```

###
## lmplot
###

```python

import seaborn as sns

# Load example dataset
tips = sns.load_dataset("tips")

# Create separate regression plots for each category of 'sex' using lmplot
sns.set_style('darkgrid')
sns.lmplot(x="total_bill", y="tip", hue="sex", data=tips)
```
