```python
import seaborn as sns
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

tips = sns.load_dataset('tips')
tips
```

###
## 1) hue:
hue emphasizes the distinction between categories through color.

###

```python
sns.relplot(x = 'total_bill', y = 'tip', data = tips, hue = 'smoker') 
```

###
## 2) style
style enhances differentiation by using different marker styles.

###

```python
sns.relplot(x = 'total_bill', y = 'tip', data = tips, style = 'smoker')
```

###
## 3) size
size enhances differentiation by using different size.

###

```python
sns.relplot(x = 'total_bill', y = 'tip', data = tips, size = 'smoker') 
```

###
##  Using all three (hue, style, size) together
###

```python
sns.relplot(x = 'total_bill', y = 'tip', data = tips, hue = 'smoker', style = 'time', size = 'size') 
```
