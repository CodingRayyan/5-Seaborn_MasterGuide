###
## Confidence Interval (remove ci = False to remove confidence level)
###

```python

import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from seaborn import relplot

fmri = sns.load_dataset('fmri')
fmri.head()
```

```python
#fig, ax = plt.subplots()
relplot( x = 'timepoint', y = 'signal', kind = 'line', data = fmri)
```

```python
relplot( x = 'timepoint', y = 'signal', kind = 'line', data = fmri, ci = False, label = 'Signal', marker = 'o')
plt.legend(loc = 'right')
```

###
## hue
###

```python

import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from seaborn import relplot

fmri = sns.load_dataset('fmri')
fmri

relplot( x = 'timepoint', y = 'signal', kind = 'line', data = fmri, hue = 'event' ,label = 'Signal', height = 4, aspect = 2,marker = 'o')
plt.legend(loc = 'right')
```

###
## style
###

```python

import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from seaborn import relplot

fmri = sns.load_dataset('fmri')
fmri

relplot( x = 'timepoint', y = 'signal', kind = 'line', data = fmri, hue = 'event' , style = 'region',label = 'Signal', height = 4, aspect = 2,marker = 'o')
plt.legend(loc = 'right')
```
